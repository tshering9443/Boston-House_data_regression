```python
# step 10 : decide the x & y values
x=df[['RM','LSTAT']].values
```

```python
y=df['MEDV']
```

```python
# BIFRUcation into train and test
# step 11: import the train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
# step 12: import the model
from sklearn.linear_model import LinearRegression
```

```python
# step 13: instantiate the model
LR = LinearRegression()
```

```python
# step 14: fit the model with train data
LR.fit(X_train,y_train)
```

```python
# step 15 :model.predicted
y_train_pred=LR.predict(X_train)
y_test_pred=LR.predict(X_test)
```

```python
# step 16:import the metrice with r2_score
from sklearn.metrics import r2_score
```

```python
train=r2_score(y_train,y_train_pred)
test=r2_score(y_test,y_test_pred) 
```

```python
print('R2_train= ',train,'\nR2_test= ',test)
```

```python
# relationship between RM & MEDV
x= df['RM'].values.reshape(-1,1)
y= df['MEDV'].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=40)
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_train_pred= LR.predict(X_train)
y_test_pred= LR.predict(X_test)
from sklearn.metrics import r2_score
train=r2_score(y_train,y_train_pred)
test=r2_score(y_test,y_test_pred)
print('R2_train= ',train,'\nR2_test= ',test)
```

```python
# Relationship between LSTATE & MEDV
x= df['LSTAT'].values.reshape(-1,1)
y= df['MEDV'].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=40)
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_train_pred= LR.predict(X_train)
y_test_pred= LR.predict(X_test)
from sklearn.metrics import r2_score
train=r2_score(y_train,y_train_pred)
test=r2_score(y_test,y_test_pred)
print('R2_train= ',train,'\nR2_test= ',test)
```

```python
# Relations with all columns
x= df.drop('MEDV',axis=1).values
y= df['MEDV'].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=40)
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
y_train_pred= LR.predict(X_train)
y_test_pred= LR.predict(X_test)
from sklearn.metrics import r2_score
train=r2_score(y_train,y_train_pred)
test=r2_score(y_test,y_test_pred)
print('R2_train= ',train,'\nR2_test= ',test)
```
